```python
movies = [
    {
        "title": "Inception",
        "ratings": [5, 4, 5, 4, 5]
    },
    {
        "title": "Interstellar",
        "ratings": [5, 5, 4, 5, 4]
    },
    {
        "title": "Dunkirk",
        "ratings": [4, 4, 4, 3, 4]
    },
    {
        "title": "The Dark Knight",
        "ratings": [5, 5, 5, 5, 5]
    },
    {
        "title": "Memento",
        "ratings": [4, 5, 4, 5, 4]
    },
]
def get_ratings(movies):
    return [sum(movie['ratings'])/len(movie['ratings']) for movie in movies]
print(get_ratings(movies))
```

```python
guests = [
    {"name": "Alice", "age": 25, "code": "VIP123"},
    {"name": "Bob", "age": 17, "code": "VIP123"},
    {"name": "Charlie", "age": 30, "code": "VIP123"},
    {"name": "Dave", "age": 22, "code": "GUEST"},
    {"name": "Eve", "age": 29, "code": "VIP123"},
]

blacklist = ["Dave", "Eve"]

PASS_CODE = "VIP123"
result = [
    guest["name"]
    for guest in guests
    if guest["age"] >= 21
    and guest["code"] == PASS_CODE
    and guest["name"] not in blacklist
]
print(result)
```

```python
val = []
for guest in guests:
    if (
        guest["age"] >= 21
        and guest["code"] == PASS_CODE
        and guest["name"] not in blacklist
    ):
        val.append(guest["name"])
print(val)
```
